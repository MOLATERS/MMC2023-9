$$W_n = (x_n+y_n)cos\alpha\\
\left \{
\begin{aligned}
x_n=\frac{sin(\frac{\theta}{2})}{sin(\frac{\pi-\theta}{2}-\alpha)}D_n \\
y_n=\frac{sin(\frac{\theta}{2})}{sin(\frac{\pi-\theta}{2}+\alpha)}D_n
\end{aligned}
\right .$$

可以推导出W_n的表达式：

$$

W_n = D_n\left(\frac{\sin\left(\frac{\theta}{2}\right)}{\sin\left(\frac{\pi - \theta}{2} - \alpha\right)} + \frac{\sin\left(\frac{\theta}{2}\right)}{\sin\left(\frac{\pi - \theta}{2} + \alpha\right)}\right) \cos \alpha

$$


```python
import numpy as np
import math
import matplotlib.pyplot as plt

theta = math.radians(120)
alpha = math.radians(1.5)
n = np.linspace(1, 5, 5)

W = (math.sin(theta)*(math.cos(alpha)**2)/(math.cos(theta/2+alpha)*math.cos(theta/2-alpha)))*(70-(n-5)*200*math.tan(alpha))


plt.plot(n, W)

plt.show()
```

```python
import matplotlib.pyplot as plt
ita = (((math.sin(theta/2))*(70-(n-4)*200*math.tan(alpha))/(math.sin(theta/2-math.pi/2-alpha))+(math.sin(theta/2))*(70-(n-5)*200*math.tan(alpha))/(math.sin(theta/2-math.pi/2+alpha)))*math.cos(alpha)-200)/W
n = np.linspace(0, 5, 5)
plt.plot(n, ita)

plt.show()
```

```python
import math 
import matplotlib.pyplot as plt
theta = math.radians(120)
alpha = math.radians(1.5)
d = 200
D = []
W = []
l = []
x = []
y = []
eta = []


for i in range(1, 10):
    l.append((i-5) * d)
    D.append(70 - (i-5) * d * math.tan(alpha))
    x.append(math.sin(theta/2) * D[i-1] / math.sin(math.pi/2-theta/2-alpha))
    y.append(math.sin(theta/2) * D[i-1] / math.sin(math.pi/2-theta/2+alpha))
    W.append(math.sin(theta) * math.cos(alpha)**2 * (70 - (i-5) * d * math.tan(alpha)) / (math.cos(theta/2-alpha) * math.cos(theta/2+alpha)))

for i in range(1, 9):
    eta.append(float(((x[i]+y[i-1]) * math.cos(alpha) - d) / W[i-1]))

plt.figure(1)
plt.plot(l, W)
plt.xlabel('$l(m)$')
plt.ylabel('$W(m)$')
plt.grid()
plt.show()
plt.figure(2)
plt.plot(l[1:], eta)
plt.xlabel('$l(m)$')
plt.ylabel('$\\eta$')
plt.grid()
plt.show()

```
